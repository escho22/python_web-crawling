```python
from bs4 import BeautifulSoup
from selenium import webdriver
import urllib.request #파일 다운로드
import urllib
import time
import sys
import math
import os
```

```python
#이미지 저장 폴더 생성
f_dir = input('이미지를 저장할 폴더를 지정하세요.(C:\\CES\\python\\notebook\\data\\images\\) : ')
query_txt = '이미지 저장'

now = time.localtime()
s = '{}-{}-{}-{}-{}-{}' .format(now.tm_year, now.tm_mon, now.tm_mday, now.tm_hour, now.tm_min, now.tm_sec)

os.makedirs(f_dir + s + '-' + query_txt) #폴더 생성
r_result_dir = f_dir + s + '-' +query_txt #확인
r_result_dir

#페이지 열기
path = 'C:/CES/python/datadown/chromedriver.exe'
driver = webdriver.Chrome(path)

s_time = time.time()

# driver.get('https://korean.visitkorea.or.kr/detail/rem_detail.do?cotid=df3f6275-06a8-4d1c-ae7e-40a2b5276bcd&temp=')
driver.get('https://korean.visitkorea.or.kr/detail/rem_detail.do?cotid=381b6381-b02a-4407-8e00-bf9a20a59111&con_type=11400')
time.sleep(1)

# #자동 스크롤 다운
# def scroll_down(driver):
#     driver.execute_script('window.scrollTo(0, document.body.scrollHeight);') #window에 요청
#     time.sleep(1)
    
# scroll_down(driver)

file_no = 0
img_src2 = []

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
img_src = soup.find('div', 'box_txtPhoto').find_all('img') #box_txtPhoto 태그 안에서 img 태그 다 가져와라

for i in img_src:
    img_src1 = i['src']
    img_src2.append(img_src1)

img_src2

#이미지 다운로드 공간 설정
os.chdir(r_result_dir)

#이미지 다운로드
for i in range(0, len(img_src2)):
    try:
        urllib.request.urlretrieve(img_src2[i], str(file_no)+'.jpg')
    except:
        continue
        
    file_no += 1
    time.sleep(0.5)
    print('{}번째 이미지 저장 중입니다......' .format(file_no))

#소요시간 및 레포트
e_time = time.time()
t_time = e_time - s_time

print('='*70)
print('총 소요시간은 {}초입니다.' .format(round(t_time, 1)))
print('총 파일 저장 건수는 {}건 입니다.' .format(file_no+1))
print('파일 저장 경로 : {}' .format(r_result_dir))
print('='*70)

#드라이버 닫기
driver.close()
```
