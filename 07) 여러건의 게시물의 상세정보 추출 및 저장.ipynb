```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
import math
import os
import random
```

```python
# 게시물 제목의 xpath
# //*[@id="listBody"]/ul/li[1]/div[2]/div[1]/a
# //*[@id="listBody"]/ul/li[2]/div[2]/div[1]/a
# //*[@id="listBody"]/ul/li[3]/div[2]/div[1]/a
```

```python
print('='*70)
print('예제 : 대한민국 구석구석 사이트의 여행지 정보 수집')
print('='*70)

query_txt = input('크롤링 키워드를 입력하세요. : ')
cnt = int(input('크롤링 최대 건수를 입력하세요. : '))
page_cnt = math.ceil(cnt/10)
```

```python
f_dir = input('결과 파일을 저장할 폴더명을 입력하세요.(예:C:\\CES\\python\\notebook\\data\\) : ')

n = time.localtime()
s = '{0}-{1}-{2}-{3}-{4}-{5}' .format(n.tm_year, n.tm_mon, n.tm_mday, n.tm_hour, n.tm_min, n.tm_sec)
os.makedirs(f_dir + s + '-' + query_txt)
ff_name = f_dir + s + '-' + query_txt + '\\' + s + '-' + query_txt + '.txt'
fc_name = f_dir + s + '-' + query_txt + '\\' + s + '-' + query_txt + '.csv'
fx_name = f_dir + s + '-' + query_txt + '\\' + s + '-' + query_txt + '.xlsx'
```

```python
s_time = time.time()
path = 'C:/CES/python/datadown/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get('https://korean.visitkorea.or.kr/')
time.sleep(random.randrange(1,5))

try:
    driver.find_element_by_xpath('//*[@id="safetyStay1"]/div[1]/div/div/button').click()
except:
    print('팝업창이 없습니다.')

element = driver.find_element_by_id('inp_search')
element.send_keys(query_txt)
element.send_keys('\n')

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

print('='*70)
print('크롤링 키워드 : ', query_txt)
print('크롤링 페이지 총 수 : ', page_cnt)
print('='*70)
```

```python
no2 =[]
contents2 = []
no = 1

for x in range(1, page_cnt+1):
    print('{} 페이지 내용 수집을 시작합니다......' .format(x))
    
    for i in range(1, 13): #중간 광고
        if no > cnt:
            break
            
        try:
            driver.find_element_by_xpath('''//*[@id="listBody"]/ul/li[{}]/div[2]/div[1]/a''' .format(i)).click()
        except:
            continue
        
        time.sleep(random.randrange(1,5))
        
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
#         content_list = soup.find('div', 'wrap_contView')

        try:
            con_1 = soup.find('div', 'txt_p').get_text()
        except:
            try:
                con_0 = soup.find('div', 'txt_p')
                con_1 = con_0.find('p').get_text()
            except:
                driver.back()
                continue
                
        print(no, ' : ', con_1)
        print('\n')
        
        f = open(ff_name, 'a', encoding='utf-8')
        f.write(str(no) + ":" + str(con_1) + '\n')
        f.close()
        
        no2.append(no)
        contents2.append(con_1)
        
        driver.back()
        time.sleep(1)
        no += 1
        
    if x > page_cnt:
        break
    
    x += 1
    
    if(x % 5 == 1) :
        driver.find_element_by_link_text('다음').click()
    else:
        driver.find_element_by_link_text('{}' .format(x)).click()
        
    time.sleep(1)
```

```python
korea = pd.DataFrame()
korea['번호'] = no2
korea['내용'] = contents2

korea.to_csv(fc_name, encoding='utf-8-sig', index=False)
korea.to_excel(fx_name, index=False)

e_time = time.time()
t_time = e_time -  s_time

print('\n')
print('='*70)
print('총 소요시간은 {}초입니다.' .format(round(t_time, 1)))
print('txt 파일 저장 완료 : {}' .format(ff_name))
print('csv 파일 저장 완료 : {}' .format(fc_name))
print('xlsx 파일 저장 완료 : {}' .format(fx_name))
print('='*70)
```
