1) melt() <br>
: 더미변수의 범주변수를 만들고 범주에 해당하는 value 변수를 만들어 정리해줌<br>
: id_vars를 제외한 나머지 고정할 열들이 variable, 그에 해당하는 행 데이터가 value 열로 정리됨<br>
: var_name, value_name 옵션으로 이름 설정 가능

```python
pew_long = pd.melt(pew, id_vars='religion')
pew_long.head()
```

```python
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')
pew_long.head()
```

2) pivot_table()<br>
: 데이터의 컬럼을 재구성해서 통계량 정리
: values => 피봇 테이블 내부에 들어갈 값
  index => 행으로 배치할 값
  columns => 열로 배치할 값
  aggfunc => calues값을 index, columns 기준으로 계산

```python
sample = pd.read_excel('../data/pandas/sample.xlsx')
sample
```

```python
sample_pivot = sample.pivot_table(values='입국객수',
                                 index = '국적명',
                                 columns = '기준년월',
                                 aggfunc = 'mean')
sample_pivot
```

```python
sample_pivot_2 = sample.pivot_table(values='입국객수',
                                   index='국적명',
                                   aggfunc='max')
sample_pivot_2
```

```python
weather = pd.read_csv('../data/pandas/weather.csv')
weather.iloc[:6, :]
```

```python
weather_melt = pd.melt(weather, 
                       id_vars=['id', 'year', 'month', 'element'],
                      var_name='day', value_name='temp')
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(values='temp',
                                       index=['id','year','month','day'],
                                       columns='element')
weather_tidy.head(10) #누락값은 자동으로 빼고
```

```python
weather_tidy_flat = weather_tidy.reset_index()
weather_tidy_flat.head(10)
```

# 중복 데이터 처리

```python
billboard = pd.read_csv('../data/pandas/billboard.csv')
billboard_long = pd.melt(billboard,
                         id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
                         var_name='week', value_name='rating')
print(billboard_long.shape)
billboard_long.head()
```

```python
billboard_long[billboard_long.track == 'Loser'].head() #track 중복데이터
```

```python
billboard_songs = billboard_long[['year', 'artist', 'track', 'time']]
billboard_songs.shape
```

```python
#drop.duplicates() : 데이터프레임의 중복데이터를 제거
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs
```

```python
billboard_ratings = billboard_long.merge(billboard_songs,
                                        on=['year', 'artist', 'track', 'time'])
billboard_ratings
```

```python
billboard_tidy = billboard_ratings.pivot_table(columns=['week'],
                                              index=['id', 'year', 'artist', 'track', 'time','date.entered'],
                                              values='rating')
billboard_tidy
```
